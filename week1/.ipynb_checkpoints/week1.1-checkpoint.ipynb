```python
import pandas
import re
```

```python
data = pandas.read_csv('titanic.csv')
data.head(5)
```

```python
# 1 Вопрос: Кол-во мужчин и женщин
output1 = open('1qw.txt', 'w')
print(data[['Sex']].groupby(['Sex']).size().male, 
         data[['Sex']].groupby(['Sex']).size().female, file=output1, end='', sep=' ')
output1.close()
```

```python
# 2 Вопрос: Какой части пассажиров удалось выжить? 
output1 = open('2qw.txt', 'w')
print(
    round(
            100*(float(data[data.Survived == 1].count().Survived)/
                float(data.shape[0])), 
            2
    ), 
    file=output1, 
    end='', 
    sep=' '
)
output1.close()
```

```python
# 3 Вопрос: Какую долю пассажиры первого класса составляли среди всех пассажиров? 

output1 = open('3qw.txt', 'w')
print(
    round(
            100*(float(data[data.Pclass == 1].count().Pclass)/
                float(data.shape[0])), 
            2

    ), 
    file=output1, 
    end='', 
    sep=' '
)
output1.close()
```

```python
# 4 Вопрос: Какого возраста были пассажиры? Посчитайте среднее и медиану возраста пассажиров

output1 = open('4qw.txt', 'w')
print(round(data.Age.mean(), 2),
      round(data.Age.median(), 2), 
    file=output1, 
    end='', 
    sep=' '
)
output1.close()
```

```python
# 5 Вопрос: Посчитайте корреляцию Пирсона между признаками SibSp и Parch.
output1 = open('5qw.txt', 'w')
print(round(data.corr().SibSp.Parch, 2), 
    file=output1, 
    end='', 
    sep=' '
)
output1.close()
```

```python
# 6 Вопрос: Какое самое популярное женское имя на корабле?

def clean_name(name):
    # Первое слово до запятой - фамилия
    s = re.search('^[^,]+, (.*)', name)
    if s:
        name = s.group(1)

    # Если есть скобки - то имя пассажира в них
    s = re.search('\(([^)]+)\)', name)
    if s:
        name = s.group(1)

    # Удаляем обращения
    name = re.sub('(Miss\. |Mrs\. |Ms\. |Mr\. |Master\. |Rev\. )', '', name)

    # Берем первое оставшееся слово и удаляем кавычки
    name = name.split(' ')[0].replace('"', '')

    return name

data['FirstName'] = data['Name'].apply(lambda x: clean_name(x))

output1 = open('6qw.txt', 'w')
print(data[data['Sex'] == 'female'][['FirstName']].groupby(['FirstName']).size().sort_values(ascending=False).head(1).index.values[0], 
    file=output1, 
    end='', 
    sep=' '
)
output1.close()
```
